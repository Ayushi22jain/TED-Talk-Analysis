```python
import numpy as np
import pandas as pd
        
df = pd.read_csv('/content/data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

Finding the most popular TED talks

```python
df.sort_values(by = 'views', ascending = False)
```

Finding the most popular TED talks Speaker (in terms of number of talks)

```python
df.groupby('author').size().sort_values(ascending = False).head()
```

 Month-wise Analysis of TED talk frequency

```python
month = []
year = []

for date in df['date']:
    month.append(date.split(' ')[0])
    year.append(date.split(' ')[1]) 
    
df['month'] = month
df['year'] = year

df.head()
```

```python
df.groupby('month').size().sort_values(ascending = False).plot(kind = 'bar')
```

Year-wise Analysis of TED talk frequency

```python
df.groupby('year').size().plot(kind = 'bar')
```

 Finding TED talks of your favorite Author

```python
df[df['author'] == 'Alex Gendler']
```

 Finding TED talks with the best view to like ratio

```python
ratio = []

data = df.values

for i in range(len(data)):
    ratio.append(data[i][3] / data[i][4])    
    
df['v&l_ratio'] = ratio

df.sort_values(by = 'v&l_ratio', ascending = False).head()
```

Finding TED talks based on tags(like education,climate,motivation etc.) 

```python
tag = 'education'

data = df.values

for i in range(len(data)):
    if (tag in data[i][0].lower()):
        print(data[i][0], data[i][5])
```

 Finding the most popular TED talks Speaker (in terms of number of views)

```python
df.groupby('author').sum()['views'].sort_values(ascending = False).head()
```
